```python
# Step 1: Define funtions to download filings
def get_list(ticker, file_type, todate):

    base_url_part1 = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK="
    base_url_part2 = "&type=&dateb=&owner=&start="
    base_url_part3 = "&count=100&output=xml"
    href = []
    
    for page_number in range(0,2000,100):
    
        base_url = base_url_part1 + ticker + base_url_part2 + str(page_number) + base_url_part3
        
        sec_page = urlopen(base_url)
        sec_soup = BeautifulSoup(sec_page)
        
        filings = sec_soup.findAll('filing')
        
        for filing in filings:
            report_year = int(filing.datefiled.get_text()[0:4])
            if (filing.type.get_text() == file_type) & (report_year > todate):
                print (filing.filinghref.get_text())
                href.append(filing.filinghref.get_text())
    
    return href

pre_url=[]
def download_report(url_list,dir_path, typefile):
    global pre_url
    target_base_url = 'http://www.sec.gov'
    
    # type = 'EX-101.INS'
    target_file_type = typefile
    
    for report_url in url_list:
        report_page = urlopen(report_url)
        report_soup = BeautifulSoup(report_page)
        
        xbrl_file = report_soup.findAll('tr')
        #print (xbrl_file)
        for item in xbrl_file:
            try:
                if item.findAll('td')[3].get_text() == target_file_type:
                    if not os.path.exists(dir_path):
                        os.makedirs(dir_path)
                    
                    end_path=item.findAll('td')[2].find('a')['href']
                    end_path=end_path.split('/')
                    #print (end_path[1])
                    if end_path[1]=='ix?doc=':
                        del end_path[1]
                    end_path=','.join(end_path)
                    end_path=end_path.replace(',', '/')
                    #print (end_path)
                    target_url = target_base_url + end_path
                    pre_url.append(target_url)
                    #print ("Target URL found!")
                    #print ("Target URL is:", target_url)
                    
                    file_name = target_url.split('/')[-1]
                    print (file_name)
                   
                    xbrl_report = urlopen(target_url)
                    output = open(os.path.join(dir_path,file_name),'wb')
                    output.write(xbrl_report.read())
                    output.close()
                    
            except:
                pass
#     return pre_url
```

```python
def text_extract_8k(url):
    from newspaper import Article

    article = Article(url)
    article.download()
    article.parse()
    text=article.text
    for i, item in enumerate (text[ :-1]):
        if item=='.':
            #print (text[i+1])
            s=text[i+1]
            try:
                s=int(s)
                text=text[ :i]+'|'+text[i+1: ]
            except:
                pass
    
    #print (text)
    return text
```

```python
def sift_tables_10Q(html):
    from bs4 import BeautifulSoup
    import csv
    html = urlopen(html)
    soup = BeautifulSoup(html)
    table = soup.findAll("table")
    #print (table)
    #output_rows = []
    output_row = []
    for a in table:
        for b in a.findAll('tr'):
            #print (b)
            for c in b.findAll('td'):
                for d in c.findAll('span'):                   
                    output_row.append(d.get_text())
#     output_row = np.array(output_row)
#     output_row=output_row.map(lambda x: [word.lower() for word in x])
    output=[]
    for str in output_row:
        output.append(str.lower())
        #print (output)
    return  output
    
# output_row
# for i, keyword in enumerate (output_row):
#     if 'Volatility' in keyword :
        
#         print (output_row[i+1])
```

```python
def find_values_10Q(hotword, file_url):
    table_list=sift_tables_10Q(file_url)
    c={}
    c_dict=[]
    #print (table_list)
    for i, keyword in enumerate (table_list):
        if hotword in keyword :
            temp=table_list[i+1]
            temp1=table_list[i+2]
            temp=temp.replace(',','')
            try:
                temp=int(temp)
                
                #print(c)
                c_dict.append(temp)
            except:
                if temp=='$':
                        #print (i)
                    temp1='$'+temp1
                    c_dict.append(temp1)
                c[hotword]=temp
    if c_dict:
        c[hotword]=c_dict

    if c=={}:
        c[hotword]=None
        
   
    return c
            
```

```python
def clean_8k (text):
    import re
    text=text.replace('(', '').replace(')','').replace('\n','')
    text=re.split(r"[;?.]", text)
    #function to remove non-ascii characters
    def removeNonAscii(s): return "".join(i for i in s if ord(i)<128)
    #remove non-ascii characters
    for t in range(len(text)):
        text[t]=removeNonAscii(text[t])
    #print (text)
    return text
```

```python
def extract_8k_values(parameter, text):
    month=['january', 'feburary','march','april','may','june','july','august','september','october','november','december']
    sentences=[]
    values=[]
    for row in text:
        row=row.lower()
        row=row.replace('|','.')
        print (row)
        if parameter in row:
            sentences.append(row)
            row=row.split(' ')
            for i, item in enumerate (row):
                item=item.replace(',','')
                try:
                    item=float(item)
                    values.append(item)
                    
                except:
                    if '%' in item or '$' in item:
                        if row[i+1]=='million':
                            item=row[i]+row[i+1]
                        values.append(item)
                    if item in month:
                        if row[i+2][0]=='2':
                            date=row[i]+row[i+1]+row[i+2]
                            values.append(date)
                        
                        
    if sentences==[]:
        sentences=None
    if values==[]:
        values=None
#     print (sentences)
#     print (values)
    return sentences, values
```

```python
def csv_8k(text,parameter_list):
    par=[]
    val=[]
    sen=[]
    for word in parameter_list:
        par.append(word)
        sent, valu=extract_8k_values(word,text)
        sen.append(sent)
        val.append(valu)
    df=pd.DataFrame(columns=['Parameters','Sentences','Values'], index=None)
    df['Parameters']=par
    df["Sentences"]=sen
    df['Values']=val
    df.to_csv('C:/data/CS/project3/par_val_8k.csv', index=None)
    print (df)
        
```
