```python
# Get AAPL stock prices with certain columns
#
data_aapl_2 = q.get_table('WIKI/PRICES', ticker = ['AAPL'],
    qopts = {'columns': ['ticker', 'date', 'adj_close', 'volume']},
    date = {'gte': '2017-12-31', 'lte': '2018-03-31' },
    paginate=True)

data_aapl_2.shape
```

```python
# Convert it to time series...
data_aapl_2['date'] = data_aapl_2['date'].astype('datetime64[s]')
data_aapl_2 = data_aapl_2.set_index('date')
data_aapl_2.info()
```

```python
# We should look at the trends and movement of the stock price via some plots
# plotting the closing price using matplotlib which is imported above 
data_aapl_2['adj_close'].plot(grid=True)

# show the plot
plt.show()
```

```python
data_aapl_2['volume'].plot(grid=True)

# show the plot
plt.show()
```

```python
# we'll be using the adjusted closing price column to calculate the percent change 
daily_close = data_aapl_2[['adj_close']]

# using pct_change method to calculate the percent change
daily_pct_c = daily_close.pct_change()

# filling the nan values with 0
daily_pct_c.fillna(0, inplace=True)

print(daily_pct_c)
```
